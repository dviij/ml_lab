```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
#question 1 
data = pd.read_csv("/home/Student/Desktop/gold_rate.csv")

# Extracting years and prices from the DataFrame
years = data["Year"].values.reshape(-1, 1)
prices = data["Price (24 karat per 10 grams)"].values

# Fitting the linear regression model
regression_model = LinearRegression()
regression_model.fit(years, prices)

# Coefficients obtained from the regression model
slope = regression_model.coef_[0]
intercept = regression_model.intercept_

print("Coefficients:")
print("Slope:", slope)
print("Intercept:", intercept)

# Predict the gold price for the year 2025
predicted_price_2025 = regression_model.predict([[2025]])
print("Predicted gold price for 2025:", predicted_price_2025[0])

# Compare with sklearn model
sklearn_model = LinearRegression().fit(years, prices)
sklearn_predicted_price_2025 = sklearn_model.predict([[2025]])

print("\nSklearn Model Coefficients:")
print("Slope:", sklearn_model.coef_[0])
print("Intercept:", sklearn_model.intercept_)

# Compute the Mean Squared Error (MSE) and Root Mean Squared Error (RMSE)
mse = mean_squared_error(prices, regression_model.predict(years))
rmse = np.sqrt(mse)

print("\nErrors:")
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
#q1
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Read gold price data from the CSV file
data = pd.read_csv("/home/Student/Desktop/gold_rate.csv")
years = data["Year"].values
prices = data["Price (24 karat per 10 grams)"].values

# Calculate manual coefficients
n = len(years)
sum_x = np.sum(years)
sum_y = np.sum(prices)
sum_xy = np.sum(years * prices)
sum_x_squared = np.sum(years ** 2)

slope_manual = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x ** 2)
intercept_manual = (sum_y - slope_manual * sum_x) / n

print("Manual Coefficients:")
print("Manual Slope:", slope_manual)
print("Manual Intercept:", intercept_manual)

# Fitting the linear regression model using sklearn
regression_model = LinearRegression()
regression_model.fit(years.reshape(-1, 1), prices)

# Coefficients obtained from the regression model using sklearn
slope_sklearn = regression_model.coef_[0]
intercept_sklearn = regression_model.intercept_

print("\nSklearn Coefficients:")
print("Sklearn Slope:", slope_sklearn)
print("Sklearn Intercept:", intercept_sklearn)

# Compare coefficients
print("\nCoefficient Comparison:")
print("Slope Match:", np.isclose(slope_manual, slope_sklearn))
print("Intercept Match:", np.isclose(intercept_manual, intercept_sklearn))

# Compute the Mean Squared Error (MSE) and Root Mean Squared Error (RMSE) using sklearn model
mse = mean_squared_error(prices, regression_model.predict(years.reshape(-1, 1)))
rmse = np.sqrt(mse)

print("\nErrors using Sklearn model:")
print("MSE:", mse)
print("RMSE:", rmse)
```

```python
# Read gold price data from the CSV file
gold_data = pd.read_csv("/home/Student/Desktop/gold_rate.csv")
gold_years = gold_data["Year"].values
gold_prices = gold_data["Price (24 karat per 10 grams)"].values

# Read silver price data from the CSV file
silver_data = pd.read_csv("/home/Student/Desktop/silver_rate.csv")
silver_years = silver_data["Year"].values
silver_prices = silver_data["Silver Rates in Rs./Kg."].values

# Merge the years and prices based on common years
common_years = np.intersect1d(gold_years, silver_years)
gold_prices = gold_prices[np.isin(gold_years, common_years)]
silver_prices = silver_prices[np.isin(silver_years, common_years)]

# Create a feature matrix with years and gold prices
feature_matrix = np.column_stack((common_years, gold_prices))

# Fitting the multiple linear regression model using sklearn
regression_model = LinearRegression()
regression_model.fit(feature_matrix, silver_prices)

# Coefficients obtained from the regression model using sklearn
coefficients_sklearn = regression_model.coef_

print("Coefficients from Sklearn:")
print("Gold Coefficient:", coefficients_sklearn[1])

# Manual calculation of coefficients using normal equation
X = feature_matrix  # Features (including the intercept term)
X_transpose = np.transpose(X)
X_transpose_X = np.dot(X_transpose, X)
X_transpose_Y = np.dot(X_transpose, silver_prices)

coefficients_manual = np.linalg.solve(X_transpose_X, X_transpose_Y)

print("\nManual Coefficients:")
print("Gold Coefficient:", coefficients_manual[1])
print("Silver Coefficient:", coefficients_manual[0])

# Predict the gold and silver prices for the year 2024
predicted_gold_price_2024 = regression_model.predict([[2024, 1]])[0]
predicted_silver_price_2024 = np.dot([2024, predicted_gold_price_2024], coefficients_manual)

print("\nPredicted gold price for 2024 (Sklearn):", predicted_gold_price_2024)
print("Predicted silver price for 2024 (Manual):", predicted_silver_price_2024)

# Compute the Mean Squared Error (MSE) and Root Mean Squared Error (RMSE) using sklearn model
mse = mean_squared_error(silver_prices, regression_model.predict(feature_matrix))
rmse = np.sqrt(mse)

print("\nErrors using Sklearn model:")
print("MSE:", mse)
print("RMSE:", rmse)
```
