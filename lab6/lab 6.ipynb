```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import SGDRegressor
```

# question 1 

```python
#question 1 
# Load the data from the CSV file
data = pd.read_csv('/home/Student/Documents/sala.csv')

# Extract the features (Experience) and target (Salary)
X = data['Experience']
Y = data['Salary']

# Calculate the means of experience and salary
X_mean = np.mean(X)
Y_mean = np.mean(Y)

# Calculate the slope (beta) and intercept (alpha) of the regression line
numerator = np.sum((X - X_mean) * (Y - Y_mean))
denominator = np.sum((X - X_mean) ** 2)

beta = numerator / denominator
alpha = Y_mean - beta * X_mean

# Create the regression line
regression_line = alpha + beta * X

# Plot the data points and the regression line
plt.scatter(X, Y, label='Data')
plt.plot(X, regression_line, color='red', label='Regression Line')
plt.title('Salary vs. Experience (Manual SLR)')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary (USD)')
plt.legend()
plt.grid(True)
plt.show()

# Print the calculated slope and intercept
print(f"Slope (beta): {beta}")
print(f"Intercept (alpha): {alpha}")
```

```python
#part1
# Load the dataset
data = pd.read_csv('/home/Student/Documents/sala.csv')

# Extract the features (X) and target (y)
X = data['Experience'].values
y = data['Salary'].values

# Initialize lists to store results
beta_values = [0.1, 1.5, 0.8]
mse_values_beta = []

# Calculate MSE for each beta value
for beta in beta_values:
    # Calculate predicted salary values using the linear regression formula
    y_pred = beta * X + 1.1 
    
    # Calculate Mean Squared Error
    mse = np.mean((y - y_pred) ** 2)
    mse_values_beta.append(mse)

# Plot Part a results
plt.figure(figsize=(8, 5))
plt.plot(beta_values, mse_values_beta, marker='o', linestyle='-')
plt.title('MSE vs. Beta (Fixed Intercept)')
plt.xlabel('Beta (Slope)')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid(True)
plt.show()
```

```python
#part 3
# Load the dataset
data = pd.read_csv('/home/Student/Documents/sala.csv')
# Extract the features (X) and target (y)
X = data['Experience'].values
y = data['Salary'].values

# Initialize lists to store results
beta_range = np.arange(0, 1.51, 0.01)
intercept_range = np.arange(0, 1.51, 0.01)
mse_values_intercept_range = np.zeros((len(beta_range), len(intercept_range)))

# Calculate MSE for each combination of beta and intercept
for i, beta in enumerate(beta_range):
    for j, intercept in enumerate(intercept_range):
        # Calculate predicted salary values using the linear regression formula
        y_pred = beta * X + intercept
        
        # Calculate Mean Squared Error
        mse = np.mean((y - y_pred) ** 2)
        mse_values_intercept_range[i, j] = mse

# Plot Part c results
plt.figure(figsize=(10, 6))
contour = plt.contourf(beta_range, intercept_range, mse_values_intercept_range, levels=50, cmap='viridis')
plt.colorbar(contour, label='Mean Squared Error (MSE)')
plt.title('MSE vs. Beta and Intercept')
plt.xlabel('Beta (Slope)')
plt.ylabel('Intercept')
plt.grid(True)
plt.show()
```

```python
#part 4
# Load the dataset
data = pd.read_csv('/home/Student/Documents/sala.csv')

# Extract the features (X) and target (y)
X = data['Experience'].values.reshape(-1, 1)
y = data['Salary'].values

# Initialize a list to store MSE values
mse_values = []

# Create a range of beta values from 0 to 1.5 with an increment of 0.01
beta_range = np.arange(0, 1.51, 0.01)

# Calculate MSE for each beta value using scikit-learn
for beta in beta_range:
    # Create and fit the SLR model
    model = LinearRegression(fit_intercept=True)
    model.coef_ = np.array([beta])
    model.intercept_ = 1.1  # Fixed intercept value
    
    # Predict using the model
    y_pred = model.predict(X)
    
    # Calculate Mean Squared Error
    mse = mean_squared_error(y, y_pred)
    mse_values.append(mse)

# Find the beta value with the lowest MSE
best_beta = beta_range[np.argmin(mse_values)]
min_mse = min(mse_values)

print(f"Best beta value: {best_beta}")
print(f"Minimum MSE: {min_mse}")
```

```python
# Load the data from the CSV file
data = pd.read_csv('/home/Student/Documents/sala.csv')

# Extract the features (Experience) and target (Salary)
X = data['Experience'].values.reshape(-1, 1)
y = data['Salary'].values

# Initialize the SGDRegressor
sgd_regressor = SGDRegressor(max_iter=5, learning_rate='constant', eta0=0.01)

# Initialize lists to store coefficients and errors
coefficients = []
errors = []

# Perform 60 iterations of 5 epochs
for iteration in range(60):
    sgd_regressor.fit(X, y)
    coefficients.append((sgd_regressor.intercept_, sgd_regressor.coef_[0]))
    y_pred = sgd_regressor.predict(X)
    mse = mean_squared_error(y, y_pred)
    errors.append(mse)

# Plot the graph of error versus iteration
plt.plot(range(60), errors)
plt.title('Error vs. Iteration (Scikit-Learn SGD)')
plt.xlabel('Iteration')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid(True)
plt.show()

# Print the final coefficients and error
final_B0, final_B1 = coefficients[-1]
final_mse = errors[-1]
print(f"Final B0: {final_B0}")
print(f"Final B1: {final_B1}")
print(f"Final MSE: {final_mse}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error

# Load the data from the CSV file
data = pd.read_csv('/home/Student/Documents/sala.csv')

# Extract the features (Experience) and target (Salary)
X = data['Experience'].values.reshape(-1, 1)
y = data['Salary'].values

# Initialize the SGDRegressor
sgd_regressor = SGDRegressor(max_iter=1, learning_rate='constant', eta0=0.01)

# Initialize lists to store coefficients and errors
coefficients = []
errors = []

# Perform multiple iterations (epochs)
epochs = 60  # You can adjust the number of epochs as needed
for epoch in range(epochs):
    sgd_regressor.fit(X, y)
    coefficients.append((sgd_regressor.intercept_, sgd_regressor.coef_[0]))
    y_pred = sgd_regressor.predict(X)
    mse = mean_squared_error(y, y_pred)
    errors.append(mse)

# Plot the graph of epoch versus error
plt.plot(range(1, epochs + 1), errors)
plt.title('Epoch vs. Error (Scikit-Learn SGD)')
plt.xlabel('Epoch')
plt.ylabel('Mean Squared Error (MSE)')
plt.grid(True)
plt.show()

# Print the final coefficients and error
final_B0, final_B1 = coefficients[-1]
final_mse = errors[-1]
print(f"Final B0: {final_B0}")
print(f"Final B1: {final_B1}")
print(f"Final MSE: {final_mse}")
```
